## Creating columns I need to calculate Adj_OE

```python
combined_wowy.head(1)
```

```python
combined_wowy['Pts_wowy_season_mins'] = combined_wowy['Pts_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
combined_wowy['Ast_wowy_season_mins'] = combined_wowy['Ast_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
combined_wowy['TO_wowy_season_mins'] = combined_wowy['TO_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
combined_wowy['3PM_wowy_season_mins'] = combined_wowy['3PM_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
combined_wowy['ORB_wowy_season_mins'] = combined_wowy['ORB_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
combined_wowy['FGA_wowy_season_mins'] = combined_wowy['FGA_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
combined_wowy['FTA_wowy_season_mins'] = combined_wowy['FTA_wowy'] / combined_wowy['Min_wowy'] * combined_wowy['MP']
```

```python
cw2 = combined_wowy.copy()
```

```python
cw2.columns
```

```python
cw2.head(1)
```

```python
cw2 = cw2[(cw2.Pos == 'SF') | (cw2.Pos == 'SG') | (cw2.Pos == 'PG')]
```

## Adj_OE_wowy

```python
cw2['Adj_OE_wowy'] = ((cw2.FGA_wowy_season_mins * cw2.FG_wowy / 100) + cw2.Ast_wowy_season_mins + (0.5 * cw2['3PM_wowy_season_mins']) + (cw2.FTA_wowy_season_mins * cw2['FT%'] * 0.44)) / (cw2.FGA_wowy_season_mins + cw2.Ast_wowy_season_mins + cw2.TO_wowy_season_mins - cw2.ORB_wowy_season_mins + (cw2.FTA_wowy_season_mins * 0.44))
```

## Creating Delta Adj OE

###  wowy Adj OE - Season Adj OE.
This means a higher OE is good because it means your Efficiency dropped less

```python
cw2['Delta_Adj_OE'] = cw2.Adj_OE_wowy - cw2.Adj_OE
```
