```python
oe_cats['FGA'] = oe_cats.FGA / oe_cats.Min
oe_cats['Pts'] = oe_cats.Pts / oe_cats.Min
oe_cats['3PA'] = oe_cats['3PA'] / oe_cats.Min
oe_cats['FTA'] = oe_cats.FTA / oe_cats.Min
oe_cats['TO'] = oe_cats.TO / oe_cats.Min
oe_cats['Ast'] = oe_cats.Ast / oe_cats.Min
oe_cats['ORB'] = oe_cats.ORB / oe_cats.Min
```

```python
oe_cats['FGA'] = oe_cats.FGA * oe_cats.MPG
oe_cats['Pts'] = oe_cats.Pts * oe_cats.MPG
oe_cats['3PA'] = oe_cats['3PA'] * oe_cats.MPG
oe_cats['FTA'] = oe_cats.FTA * oe_cats.MPG
oe_cats['TO'] = oe_cats.TO * oe_cats.MPG
oe_cats['Ast'] = oe_cats.Ast * oe_cats.MPG
oe_cats['ORB'] = oe_cats.ORB * oe_cats.MPG
```

```python
def add_cols(team):
    team['FGM'] = (team.FGA * team.FG / 100.0)
    team['3PM'] = (team['3PA'] * team['3P%'] / 100.0)
    team['FTM'] = (team['FTA'] * team['FT%'] / 100.0)

## calculating oe of the team
def calc_oe(team):
    team = team.sum()
    team_oe = (team.FGM + team.Ast + (0.5 * (team['3PM']) + (team.FTM * 0.44))) / (team.FGA + team.Ast + team.TO - team.ORB + (team.FTA * 0.44))
    return team_oe
```

```python
driver = webdriver.Chrome()
sleep(2)
driver.get("http://stats.nba.com/players/speed-distance/#!?Season=2013-14&SeasonType=Regular%20Season")
sleep(2)
driver.set_page_load_timeout(30)
while True:
    try:
        driver.find_element_by_class_name('table-addrows__button').click()
    except:
        break
html2 = driver.page_source
```

```python
soup2 = BeautifulSoup(html2)
```

```python
distance = pd.DataFrame(columns=['Name','Team','GP','Mins','Total Miles Run','Miles Run O','Avg MPH Run O'])
```

```python
for i in soup2.find_all('div',{'class':'nba-stat-table__overflow'}):
    for j in i.find_all('tr',{'class':'ng-scope'}):
        row = j.text
        chop = row.split('\n')
        print chop
        name = chop[1]
        team = chop[2]
        gp = chop[3]
        mins = chop[6]
        total = chop[8]
        miles_offense = chop[9]
        speed_off = chop[12]
        distance.loc[len(distance)] = [name,team,gp,mins,total,miles_offense,speed_off]
```

```python
distance.tail()
```

```python
distance['Mins'] = distance['Mins'].astype(float)
distance['GP'] = distance['GP'].astype(int)
distance['Total Miles Run'] = distance['Total Miles Run'].astype(float)
distance['Miles Run O'] = distance['Miles Run O'].astype(float)
distance['Avg MPH Run O'] = distance['Avg MPH Run O'].astype(float)
```

```python
distance['Miles Run Per 36'] = distance['Total Miles Run'] * 36 / distance.Mins
distance['Miles Run O Per 36'] = distance['Miles Run O'] * 36 / distance.Mins
```

```python
drive['PTS Per Drive'] = drive['Points Driving'] / drive['Num Drives']
drive['Assists Per Drive'] = drive['Assists Driving'] / drive['Num Drives']
drive['TO Per Drive'] = drive['Turnovers Driving'] / drive['Num Drives']
```

## Merging NBA.com dfs

```python
contested = pd.read_csv('basketball_csvs/contested_2014.csv')
distance = pd.read_csv('basketball_csvs/distance_2014.csv')
drives = pd.read_csv('basketball_csvs/drives_2014.csv')
finishing = pd.read_csv('basketball_csvs/finishing_2014.csv')
pullups = pd.read_csv('basketball_csvs/pullups_2014.csv')
shot_off_dribble = pd.read_csv('basketball_csvs/shot_off_dribble_2014.csv')
scoring_locations = pd.read_csv('basketball_csvs/scoring_locations_2014.csv')
```

```python
all_nba= pd.DataFrame()
```

```python
all_nba = distance.merge(drives, on='Name', how='outer')
```

```python
all_nba = all_nba.merge(finishing, on='Name', how='outer')
```

```python
all_nba = all_nba.merge(contested, on='Name', how='outer')
```

```python
all_nba = all_nba.merge(pullups, on='Name', how='outer')
```

```python
all_nba = all_nba.merge(shot_off_dribble, on='Name', how='outer')
```

```python
all_nba = all_nba.merge(scoring_locations, on='Name', how='outer')
```
